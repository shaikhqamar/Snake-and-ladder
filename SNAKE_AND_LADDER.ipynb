```python
import random as r # imported random library with alias name 'r' to generate random numbers for the dice

# defined function for the ladder and  snake condition to move the player,s position accordingly
def ladder_or_snake(position,playername): # given 2 parameter postion for player position, playername for the player's name
    if position in board: 
        if board[position][0] == 'ladder':  
            print("wooo!! {} got a ladder ".format(playername))
            position = board[position][1] 
            print("{} climbed at {}".format(playername, position))
        elif board[position][0] == 's-head':  
            print("oooppss!! Snake encounter ")
            position = board[position][1]  
            print("{} fell at {}".format(playername, position))
    return position        

# defined a functon for player movement according to the value of dice 
def moves(dice,position,playername):# given 3 parameteras dice for the dice value, postion for player position, playername for the player's name
    if (position + dice) <= 100:
        position += dice
        position = ladder_or_snake(position,playername)
        print("{} is at {} now".format(playername, position))
    else:
        position = position
        position = ladder_or_snake(position,playername)
        print("{} is remained at {}".format(playername, position)) 
    return position

# defined a function to start the game 
def play(position,playername):# given 2 parameter postion for player position, playername for the player's name
    dice = r.choice(range(1, 7)) # picked a random number from given range and store it in variable dice
    print("Dice got :", dice) 
    
    # if the dice value is six
    if dice == 6:
        position = moves(dice,position,playername) # function of player movement
        # if player got a six then for the extra move it again ask to the same player to roll
        input("{} got 6 press enter to roll again :".format(playername))
        dice = r.choice(range(1, 7)) 
        print("Dice got :", dice)
        if dice == 6:
            position = moves(dice,position,playername) # function of player movement
        else:
            position = moves(dice,position,playername) # function of player movement
    #  if the dice value isn't six
    else:
        position = moves(dice,position,playername) # function of player movement
    return position


# created a dictionary to check the snake-bite position and ladder , and jump the player postion accordingly 
board = {1: ['ladder', 38], 4: ['ladder', 14], 8: ['ladder', 30], 21: ['ladder', 42], 28: ['ladder', 76],
         32: ['s-head', 10], 36: ['s-head', 6], 48: ['s-head', 26], 50: ['ladder', 67], 62: ['s-head', 18],
         71: ['ladder', 92], 80: ['ladder', 99], 88: ['s-head', 24], 95: ['s-head', 56], 97: ['s-head', 78]}

player1_pos = 0 #80
player2_pos = 0 #80
player1_name = input("Player1 name :") # 1st player name
player2_name = input("Player2 name :") # 2nd player name
while (player1_pos or player2_pos) != 100: # condition to run the loop until the it is false
        player1 = int(input("{} Press 1 to roll the dice : ".format(player1_name)))
        player1_pos = play(player1_pos,player1_name)
        if player1_pos==100:
            print(f"GAME---OVER\n CONGRATULATION {player1_name} won the game")
            break
    
        player2 = int(input("{} Press 2 to roll the dice :  ".format(player2_name)))
        player2_pos = play(player2_pos,player2_name)
        if player2_pos==100:
            print(f"GAME---OVER\n CONGRATULATION {player2_name} won the game")
            break
```
